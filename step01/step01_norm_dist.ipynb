### 정규 분포

```python
def normal(x, mu = 0, sigma = 1): # 평균 0, 표준편차 1인 정규분포의 확률 밀도 함수
  y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-((x - mu) ** 2) / (2 * (sigma ** 2)))
  return y
```

```python
x = np.linspace(-5, 5, 100) # -5 ~ 5 범위의 임의의 숫자 100개 랜덤 생성
y = normal(x)

# 정규분포 시각화 하기 -> 종 모양 곡선 생성
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 평균이 달라질 경우, 정규분포 그래프의 폭은 같지만 확률밀도가 최대가 되는 위치가 달라짐
x = np.linspace(-10, 10, 1000)

y0 = normal(x, mu = -3)
y1 = normal(x, mu = 0)
y2 = normal(x, mu = 5)

plt.plot(x, y0, label = '$\mu$=-3')
plt.plot(x, y1, label = '$\mu$=0')
plt.plot(x, y2, label = '$\mu$=5')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 표준편차의 변화에 따라 정규분포 폭이 달라짐. 표준편차가 클수록 산의 모양이 낮고 넓으며, 작을수록 높고 좁음

y0 = normal(x, mu = 0, sigma = 0.5)
y1 = normal(x, mu = 0, sigma = 1)
y2 = normal(x, mu = 0, sigma = 2)

plt.plot(x, y0, label = '$\sigma$=0.5')
plt.plot(x, y1, label = '$\sigma$=1')
plt.plot(x, y2, label = '$\sigma$=2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 중심 극한 정리 실험

- 샘플의 크기가 커질수록 표본평균의 분포가 점점 정규분포를 따른다는 것을 알 수 있음

```python
N = 3 # 샘플 크기

xs = []
for n in range(N):
  x = np.random.rand() # 균등 분포로부터 무작위 샘플링
  xs.append(x)

x_mean = np.mean(xs)
print(x_mean)
```

```python
# N = 1 인 경우 10000개의 샘플(샘플은 관측값의 집합(관측값은 특정 확률 분포를 띠는 1개의 값!))에 대한 표본 평균의 분포 구하기
x_means = []
N = 1

for _ in range(10000):
  xs = []
  for n in range(N):
    x = np.random.rand()
    xs.append(x)
  mean = np.mean(xs)
  x_means.append(mean)

plt.hist(x_means, bins = 'auto', density=True)
plt.title(f'N={N}')
plt.xlabel('x')
plt.ylabel('probability density')
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

### 표본 합의 확률분포

#### 균등 분포
- 평균: 1/2
- 분산 : 1/12
따라서 샘플 크기가 N인 표본 합의 분포는 평균이 N/2이고 분산이 N/12인 정규 분포에 가까워짐

```python
x_sums = []
N = 5

for _ in range(10000):
  xs = []
  for n in range(N):
    x = np.random.rand()
    xs.append(x)
  t = np.sum(xs)
  x_sums.append(t)

x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)

plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.title(f'N={N}')
plt.xlim(-1, 6)
plt.show()
```
